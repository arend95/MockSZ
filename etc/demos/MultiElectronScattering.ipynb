# Photon scattering from an electron population

Now that we have discussed the single electron-photon scattering scenario, it is time to discuss the case of photon scattering. The difference is that instead of a single $\beta$, an electron population contains a whole range of $\beta$'s. This is reflected by including the electron distribution in the total probability. It should be noted that each scattering probability shown here is for a single scattering only. This amounts to us setting $\tau_\mathrm{e}=1$.

We start by showing scattering probabilities for a relativistic Maxwellian distributions of electrons.

```python
import numpy as np
import matplotlib.pyplot as pt

import MockSZ.MultiStats as MMulti
import MockSZ.Conversions as MConv

Te = MConv.eV_To_Temp(np.array([1, 3, 9]) * 5.1e3)
num_arr = 1000

s = np.linspace(-1.4, 1.4, num=num_arr)
   
amplitudes0 = MMulti.getP1_RM(s, Te[0])
amplitudes1 = MMulti.getP1_RM(s, Te[1])
amplitudes2 = MMulti.getP1_RM(s, Te[2])

fig, ax = pt.subplots(1,1)
ax.plot(s, amplitudes0, label=r"$T_\mathrm{e}$ = 5.1 KeV")
ax.plot(s, amplitudes1, label=r"$T_\mathrm{e}$ = 15.3 KeV")
ax.plot(s, amplitudes2, label=r"$T_\mathrm{e}$ = 45.9 KeV")
ax.set_ylim(0, 5)
ax.set_xlim(-1.4, 1.4)

ax.set_ylabel(r"$P_1(s)$")
ax.set_xlabel(r"$s$")
ax.legend(frameon=False, prop={'size': 13},handlelength=1)

pt.show()
```

It can be seen that the higher the electron temperature $T_\mathrm{e}$, the broader the distribution becomes. This is expected from the Maxwellian distribution, which gets broader with increasing $T_\mathrm{e}$.

The second scattering probabilities we will inspect belong to a non-thermal power law distribution. This type is often used for modelling non-thermalised electron populations.

```python
alpha = np.array([2, 2.5, 3])
s = np.linspace(-1.4, 10, num=num_arr)

amplitudes0 = MMulti.getP1_PL(s, alpha[0])
amplitudes1 = MMulti.getP1_PL(s, alpha[1])
amplitudes2 = MMulti.getP1_PL(s, alpha[2])

fig, ax = pt.subplots(1,1)
ax.plot(s, amplitudes0, label=r"$\alpha$ = 2")
ax.plot(s, amplitudes1, label=r"$\alpha$ = 2.5")
ax.plot(s, amplitudes2, label=r"$\alpha$ = 3")
ax.set_ylim(0, 0.8)
ax.set_xlim(-1.4, 10)

ax.set_ylabel(r"$P_1(s)$")
ax.set_xlabel(r"$s$")
ax.legend(frameon=False, prop={'size': 13},handlelength=1)

pt.show()
```

It can immediately be seen that the high-frequency scattering tail is a lot longer compared to the thermal Maxwellian. This is due to the larger population of electrons at higher velocities, which translates into larger frequency shifts.
